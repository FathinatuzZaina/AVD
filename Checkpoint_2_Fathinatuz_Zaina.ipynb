# **DATA PREPARATION**

MENCARI MISSING VALUES

```python
(df.isna().sum()/len(df))*100
```

MENCARI DATA SAMA (DUPLIKAT)

```python
df[df.duplicated()]
```

OUTLIERS

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - (1.5 * iqr)
  upper_bound = q3 + (1.5 * iqr)
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers) / len(df)) * 100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

display(results_df)
```

IMPUTASI

```python
columns_to_impute = [
    "Protein-g", "Fiber-g", "Sodium-g",
    "SaturatedFat-g", "MonounsaturatedFat-g",
    "PolyunsaturatedFat-g", "TransaFat-g",
    "MeasureQuantity", "MQGmWt"
]

for col in columns_to_impute:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    df[col] = df[col].clip(lower=lower_bound, upper=upper_bound)
```

MEMERIKSA KEMBALI

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - (1.5 * iqr)
  upper_bound = q3 + (1.5 * iqr)
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers) / len(df)) * 100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

display(results_df)
```

MEMERIKSA INCONSISTEN VALUE

```python
inconsistent_types = df[~df['Calories-kcl'].apply(lambda x: isinstance(x, int)) |
                         ~df['Sodium-g'].apply(lambda x: isinstance(x, int)) |
                         ~df['MQCalories-kcl'].apply(lambda x: isinstance(x, int)) |
                         ~df['MQCalories100gm-kcl'].apply(lambda x: isinstance(x, int)) |
                         ~df['Protein-g'].apply(lambda x: isinstance(x, float)) |
                         ~df['Carb-g'].apply(lambda x: isinstance(x, float)) |
                         ~df['Fiber-g'].apply(lambda x: isinstance(x, float)) |
                         ~df['Sugar-g'].apply(lambda x: isinstance(x, float)) |
                         ~df['SaturatedFat-g'].apply(lambda x: isinstance(x, float)) |
                         ~df['MonounsaturatedFat-g'].apply(lambda x: isinstance(x, float)) |
                         ~df['PolyunsaturatedFat-g'].apply(lambda x: isinstance(x, float)) |
                         ~df['TransaFat-g'].apply(lambda x: isinstance(x, float)) |
                         ~df['MeasureQuantity'].apply(lambda x: isinstance(x, float)) |
                         ~df['MQGmWt'].apply(lambda x: isinstance(x, float))]

inconsistent_scores = df[(df['Calories-kcl'] < 0) |
                          (df['Sodium-g'] < 0) |
                          (df['Protein-g'] < 0) |
                          (df['Carb-g'] < 0) |
                          (df['Fiber-g'] < 0) |
                          (df['Sugar-g'] < 0) |
                          (df['SaturatedFat-g'] < 0) |
                          (df['MonounsaturatedFat-g'] < 0) |
                          (df['PolyunsaturatedFat-g'] < 0) |
                          (df['TransaFat-g'] < 0) |
                          (df['MeasureQuantity'] < 0) |
                          (df['MQCalories-kcl'] < 0) |
                          (df['MQCalories100gm-kcl'] < 0)]

inconsistent_categories = df[~df['ProductType'].isin(['Type1', 'Type2', 'Type3']) |  # Ganti dengan kategori yang sesuai
                              ~df['MeasureType'].isin(['grams', 'ml', 'pieces'])]  # Ganti dengan kategori yang sesuai

inconsistent_values = pd.concat([inconsistent_types, inconsistent_scores, inconsistent_categories]).drop_duplicates()

print(inconsistent_values)
```

CONSTRUCT DATA

```python
def sehat_semua(row):
    batas_sehat_calories = 200
    batas_sehat_protein = 5

    if row['Calories-kcl'] <= batas_sehat_calories and row['Protein-g'] >= batas_sehat_protein:
        return 'Sehat'
    else:
        return 'Tidak Sehat'

df['Keterangan'] = df.apply(sehat_semua, axis=1)

print(df[['Food', 'ProductType', 'Calories-kcl', 'Protein-g', 'Keterangan']].head(10))
```
